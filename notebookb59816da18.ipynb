```python
import numpy as np
import pandas as pd
import matplotlib as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
```

```python
data.head()
```

```python
data_features = ['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking','time']
data_pred_features = ['DEATH_EVENT']
features = data[data_features]
preds = data[data_pred_features]
```

```python
feature_scaler = StandardScaler()
feature_scaler.fit(features)
features = feature_scaler.transform(features)
```

```python
features_train, features_test, preds_train, preds_test = train_test_split(features, preds, test_size=0.3)
```

```python
features_train, features_validate, preds_train, preds_validate = train_test_split(features_train,preds_train, test_size=0.2)
```

```python
input_layer = tf.keras.Input(shape=(12,))
x = tf.keras.layers.Dense(12,activation=tf.nn.sigmoid)(input_layer)
x2 = tf.keras.layers.Dense(24,activation=tf.nn.sigmoid)(x)
x3 = tf.keras.layers.Dense(12, activation=tf.nn.sigmoid)(x2)
outputs = tf.keras.layers.Dense(1,activation=tf.nn.sigmoid)(x3)
model = tf.keras.Model(inputs=input_layer, outputs=outputs)
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy")
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)
model.fit(features_train, preds_train, validation_data=(features_validate,preds_validate), epochs=1000, callbacks=[callback])
```

```python
model.evaluate(features_test,preds_test)
```

```python
model.predict(features_test).round()
```
